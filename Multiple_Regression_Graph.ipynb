```python
X1,y1=df1['Price of House(in million)'].to_numpy().reshape(-1,1),df1['Water availability (in Hours)'].to_numpy().reshape(-1,1)
```

```python
X,y=df1['Water availability (in Hours)'].to_numpy().reshape(-1,1),df1['Price of House(in million)'].to_numpy().reshape(-1,1)

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)

from sklearn.linear_model import LinearRegression
linreg=LinearRegression().fit(X_train,y_train)
print(linreg.coef_,linreg.intercept_)

plt.figure(figsize=(10,5))
plt.scatter(y=df['Price of House(in million)'],x=df['Water availability (in Hours)'])
plt.title('Price of House VS Water Availibility', fontsize=25)
plt.plot(df['Water availability (in Hours)'],linreg.coef_*X+linreg.intercept_,color='red')
plt.plot()
```

```python
X,y=df2["Electricity Availability (in Hours)"].to_numpy().reshape(-1,1),df2['Price of House(in million)'].to_numpy().reshape(-1,1)

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)

from sklearn.linear_model import LinearRegression
linreg=LinearRegression().fit(X_train,y_train)
print(linreg.coef_,linreg.intercept_)

plt.figure(figsize=(10,5))
plt.scatter(y=df2['Price of House(in million)'],x=df2["Electricity Availability (in Hours)"])
plt.title('Price of House VS Electricity Availability', fontsize=25)
plt.plot(df2["Electricity Availability (in Hours)"],linreg.coef_*X+linreg.intercept_,color='red')
plt.plot()
```

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
 
 
# Creating dataset
x = df.iloc[:,1]
y = df.iloc[:,2]
z =  df.iloc[:,0]
 
# Creating figure
#fig = plt.figure(figsize = (10, 7))
fig=plt.figure(figsize=(10,8))
ax = plt.axes(projection ="3d")
 
# Creating plot
ax = fig.add_subplot(111,projection='3d')


xx=x.to_numpy()
yy=y.to_numpy()


X,Y = np.meshgrid(xx, yy)
eq = 0.6171662125340589 + 0.19264305*X + 0.22561308*Y



ax.plot_surface(X, Y, eq,alpha=0.1,color="lightgreen")
ax.scatter3D(x, y, z, color = "black")
plt.title("simple 3D scatter plot")
ax.set_xlabel('X-axis : Water Availability', fontweight ='bold')
ax.set_ylabel('Y-axis : Electricity AVailability', fontweight ='bold')
ax.set_zlabel('Z-axis : House Price', fontweight ='bold')
# show plot
plt.show()
```
